# NLP  :- Nature lanagage Process 

```python
df['tags'][0]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmarizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\sa-zA-Z0-9]', '',text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemmarizer.lemmatize(word) for word in words] 
    return " ".join(words)
    
    
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df.head()
```

```python
df['tags'][0]
```

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

# Creating Model

```python
from  sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommed(title, n = 10):
    if title not in indices: 
        return ["Movie not found"]

    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
```

```python
recommed('Toy Story')
```

```python
recommed('Avenger')
```

```python
recommed('Avatar')
```
